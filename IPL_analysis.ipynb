# Exploratory Data Analysis

```python
sns.countplot(data=df,x='season')
sns.set(rc={'figure.figsize':(15,5)})
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax=sns.countplot(df, x = 'winner' ,order=df['winner'].value_counts().index)

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count of the most winner in the matches")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax=sns.countplot(df, x= 'toss_winner' ,order=df['toss_winner'].value_counts().index)

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count of most toss winner ")
plt.xticks(rotation =90)
plt.show()

```

```python
ax=sns.countplot(df, x= 'toss_decision')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count of toss decision ")

plt.show()

```

```python
ax=sns.countplot(df, x= 'city')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation = 90)
plt.show()
```

```python
df.head(3)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
grouped = df.groupby(['winner', 'toss_winner', 'toss_decision']).sum().reset_index()
ax=sns.histplot(data=grouped, x='winner', hue='toss_decision', multiple='dodge')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation =90)
plt.show()
```

```python
df.dtypes
```

```python
corr_matrix = df[['id', 'season', 'win_by_runs', 'win_by_wickets']].corr()
sns.heatmap(corr_matrix, cmap='YlGnBu', annot=True)
```

```python
top_players = df['player_of_match'].value_counts().nlargest(10)
ax=sns.barplot(x=top_players.index, y=top_players.values)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Top 10 player_of_match in IPL')
plt.xlabel('Player')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
first_bat_win_team = df[df['win_by_runs']>0]
top_teams = first_bat_win_team['winner'].value_counts().nlargest(3)
sns.barplot(x=top_teams.index , y =top_teams.values)
plt.title('Top 3 teams who bat first and win in IPL')
plt.show()
```

```python
first_ball_win_team = df[df['win_by_wickets']>0]
top_teams = first_ball_win_team['winner'].value_counts().nlargest(3)
sns.barplot(x=top_teams.index , y =top_teams.values)
plt.title('Top 3 teams who ball first and win in IPL')
plt.show()
```
